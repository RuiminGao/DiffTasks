```python
# average for each combination of Subject_Type, Task_Type
summary_lh = data_lh.groupby(['Subject_Type', 'Task_Type'])['DC'].mean().reset_index()
summary_rh = data_rh.groupby(['Subject_Type', 'Task_Type'])['DC'].mean().reset_index()
```
