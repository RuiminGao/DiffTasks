```python
import pandas as pd
data = pd.read_csv('../../DiffTasks_subject/Subjects.csv')
difftasks_subs = pd.read_csv('qc/sessions4toolbox.csv')['UID']
data_difftasks = data[data['Subjects::UniqueID'].isin(difftasks_subs)]
data_difftasks = data_difftasks[data_difftasks['Experiment'].str.contains('DiffTasks')]
data_difftasks = data_difftasks.drop_duplicates(subset='Subjects::UniqueID', keep='last')
data_difftasks = data_difftasks.sort_values(by='Subjects::UniqueID')
```
