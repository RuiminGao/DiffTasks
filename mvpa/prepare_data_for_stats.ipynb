```python
import pandas as pd
import ast
import numpy as np
```

```python
systems = ['Lang', 'MD', 'Lang_RH']
configs = [
    ('conditions', 'Stimulus', 'Same'),
    ('tasks', 'Task', 'Same'),
    ('conditions', 'Stimulus', 'Diff'),
    ('tasks', 'Task', 'Diff'),
]

data = []
for system in systems:
    system_name = 'Language_LH' if system == 'Lang' else 'Language_RH' if system == 'Lang_RH' else system
    for cond_type, target, train_type in configs:
        fname = f'Data/mvpa_{system}_{cond_type}_{"same" if train_type=="Same" else "cross"}_pattern_details.csv'
        dat = pd.read_csv(fname).reset_index(drop=True)
        dat['TrainingType'] = train_type
        dat['System'] = system_name
        dat['Target'] = target
        data.append(dat)

data = pd.concat(data)
data = data.rename(columns={'Tasks': 'TestingTask', 'Conditions': 'TestingStimulus'})
data['Y_True'] = data['Y_True'].apply(ast.literal_eval)
data['Y_Pred'] = data['Y_Pred'].apply(ast.literal_eval)
data['Y_Pred'] = data['Y_Pred'].apply(lambda x: 
                                      [char for char in x[0]] if len(x) == 1 and not x[0].startswith('V')  
                                      else [x[0][i:i+2] for i in range(0, len(x[0]), 2)] if len(x) == 1 and x[0].startswith('V')
                                      else x)
data = data.explode(['Y_True', 'Y_Pred']).reset_index(drop=True)
```
